```python
import pandas as pd
import requests

# --- CONFIGURATION ---
GITHUB_TOKEN = 'your github token here - do not share it'
OWNER = 'Abson-dev'
REPO = 'Projet-Statistique-sur-R-et-Python-2025-2026'
EXCEL_FILE = 'Cours de projet statistique ISE 1 2025-2026.xlsx'
COLUMN_NAME = 'Github_User'  # The column name in your Excel file

headers = {
    'Authorization': f'token {GITHUB_TOKEN}',
    'Accept': 'application/vnd.github.v3+json'
}

def invite_student(username):
    url = f'https://api.github.com/repos/{OWNER}/{REPO}/collaborators/{username}'
    # Permission can be 'pull', 'triage', 'push', 'maintain', or 'admin'
    data = {'permission': 'push'}

    response = requests.put(url, headers=headers, json=data)

    if response.status_code == 201:
        print(f"✅ Invitation sent to {username}")
    elif response.status_code == 204:
        print(f"ℹ️ {username} is already a collaborator.")
    else:
        print(f"❌ Failed to invite {username}: {response.json().get('message')}")

# --- EXECUTION ---
try:
    df = pd.read_excel(EXCEL_FILE)
    students = df[COLUMN_NAME].dropna().tolist()

    for student in students:
        invite_student(student.strip())
except Exception as e:
    print(f"Error reading file: {e}")
```

```python
def check_pending_invitations():
    url = f'https://api.github.com/repos/{OWNER}/{REPO}/invitations'
    response = requests.get(url, headers=headers)

    if response.status_code == 200:
        invites = response.json()
        if not invites:
            print("🙌 No pending invitations. Everyone has accepted!")
        else:
            print(f"⏳ Found {len(invites)} pending invitations:")
            for invite in invites:
                # 'invitee' contains the user details
                username = invite['invitee']['login']
                invite_date = invite['created_at']
                print(f" - {username} (Sent on: {invite_date})")
    else:
        print(f"❌ Could not retrieve invitations: {response.status_code}")

# Call this after your invitation loop or as a standalone check
check_pending_invitations()
```
